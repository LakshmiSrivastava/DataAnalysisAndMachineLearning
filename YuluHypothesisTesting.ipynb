**Hypothesis Testing**

**1.If working day has an effect on the number of electric cycles rented**

```python
from scipy.stats.mstats import pearsonr
print('Ho : workingday and count of renting are independent ')
print('Ha : workingday and count of renting are dependent\n')

print("Mean value of number of ebikes renting depend on working day")
print(df1.groupby('workingday')['count'].mean())
print()
df_wd=df1[df1['workingday']=='working']['count'].values
df_nwd=df1[df1['workingday']=='No_working']['count'].values
print(f'Individual mean of working day bikes renting count {df_wd.mean()}')
print(f'Individual mean of Non working day bikes renting count {df_nwd.mean()}')
print()
print('visualization to get whether the plot is normal or not ')
fig, ax = plt.subplots(2,2, figsize=(15, 5))
san.kdeplot(df_wd, ax=ax[0][0])
san.kdeplot(df_nwd, ax=ax[0][1])
qqplot(df_wd, line='r', ax=ax[1][0])
qqplot(df_nwd, line='r',ax=ax[1][1])
plt.show()
print("qqplot shows that the distribution is normal")

print('Covarriance test ')
var_wd=np.var(df_wd)
var_nwd=np.var(df_nwd)
print()
print(f'Variance of workingday {var_wd} and variance of non working day , {var_nwd}')
print("Variance of the data groups is less than 4:1, variance are same")
print("ttest on data group to check dependancy ")
print()
print("The alpha value is 0.05")
ttest_stat, p_value=ttest_ind(df_wd, df_nwd, alternative='less')
alpha=0.05
print("p_value", p_value)
if p_value<0.05:
  print("Ho rejected")
  print("Workingday and count of renting is dependent")
else:
  print("Ho is fail to reject")
  print("Workingday and count of renting is independent")

```

**Insight:**

Kdplot shows that the distribution is log normal between count of renting and workingday.

And QQ plot shows that there is minimum deviation from regression line but not a normal distribution.

Calculation of Covariance between these group are 34040.70 and 30171.35 and there ratio is less than 4:1, so ttest can perform .

The p_value of ttest is 0.8867759788681933 which is greater than 0.05.

**Recommendation:** On the basis of these statustically proved data , we can say that Ho is fail to reject ie we dont have sufficient evidance to say that working day has an effect on the number of count of renting.

**3. Hypothesis testing : Weather is dependent on season**

```python
print("The visualization of weather is depend on season is shown in bivariate analysis between weather and season")
print("The result shows that during fall, most of the time weather is clear and winter - misty")
df_w=df['weather']
df_s=df['season']
print()
print("calculating means of individual group ")
print(" mean of weather : " ,df_w.mean())
print(" mean of season : " ,df_s.mean())
print()
print("Perform ttest to check the means of groups are different?")
tstat, p_valuet=ttest_ind(df_w, df_s)
print(" Tstat value :", tstat,"  &  ", "p_value of Ttest : ", p_valuet)
if p_valuet<0.05:
  print("The means of two groups are different")
else:
  print("The means of two groups are same")
print()
print(" Check if the distribution are different")
qqplot(df_w, line='r')
qqplot(df_s, line='r')
plt.show()
print("qqplot shows that the distribution is not normal")

kstat, p_valuek=kstest(df_w, df_s)
print(" kstat value :", kstat,"  &  ", "p_value of kstest : ", p_valuek)
if p_valuek<0.05:
  print("The distribution is not normal")
else:
  print("the distribution is normal")
print()
print(" Both are categorical data, we can perform chisquare test for dependency")
contigency=pd.crosstab(df['weather'],df['season'])
chistat, p_valuechi,dof, x=chi2_contingency(contigency)
print(" Chi2stat value :", chistat,"  &  ", "p_value of chi2 test : ", p_valuechi)
if p_valuechi<0.05:
  print("The weather and season are dependent")
else:
  print("The weather and season are independent")
```

3. Statically proved it by ttest.
4. then  checked for the distribution , qqplot shows that distribution are not normal.
5. Checked for the dependency by chi2 test. Result shows statistical significance between weather and season showed.

1. There is a statistical significanse dependency between weather and season, So wether prediction with varying season should keep in consideration. As weather depends on season, so have ample time to take decision because already know the weather and its effect on count of renting.
