Handling Null values

Insight: There are 2634 null values in director column, 825 in cast, 831 in country, 10 in date_added, 4 in rating and 3 in duration.

So imputing null values of duration with 0,country with 'United states' as aired by US mostly, cast also with 'Anonymous', director with 'Anonymous', rating with 'Not Given' , and date added with previous value.

```python
nf1['duration']=nf1['duration'].fillna(0,axis=0)
nf1['country']=nf1['country'].fillna('United States',axis=0)
nf1['cast']=nf1['cast'].fillna('Anonymous',axis=0)
nf1['director']=nf1['director'].fillna('Anonymous',axis=0)
nf1['rating']=nf1['rating'].fillna('Not Given',axis=0)
nf1['date_added']=nf1['date_added'].interpolate(method='ffill')
```

```python
p_25=np.percentile(nf['release_year'],25)
p_50=np.percentile(nf['release_year'],50)
p_75=np.percentile(nf['release_year'],75)
iqr=p_75-p_25
upper_limit=p_75+(1.5*iqr)
lower_limit=max(p_25-(1.5*iqr),0)
max_year=nf['release_year'].max()
min_year=nf['release_year'].min()
print("Minimum Year", min_year)
print("Maximum Year", max_year)
print("First Quartile(Q1)", p_25)
print("Second Quartile(Q2)", p_50)
print("Third Quartile", p_75)
print("IQR", iqr)
print("Upper Limit", upper_limit)
print("Lower Limit", lower_limit)
sns.boxplot(data=nf['release_year'], orient='h')
print("Totl value", len(nf))
Year_outlier=nf[(nf['release_year']<lower_limit) | (nf['release_year']>upper_limit)]
print("Total outlier", len(Year_outlier))
percent_year_outlier=len(Year_outlier)/len(nf)
print("Percent of outlier", percent_year_outlier)

```
