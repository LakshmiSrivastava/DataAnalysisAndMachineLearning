CLT Analysis for mean purchase with confidence 95% based on Gender for sample size (100,1000,1000)

```python
def bootstrapMvsF(sam1, sam2, samSize=500, itrSize=5000, CL=0.95, NumTails=2):
  sam1Mean=np.empty(itrSize)
  sam2Mean=np.empty(itrSize)
  for i in range(itrSize):
    sam1N=np.empty(samSize)
    sam2N=np.empty(samSize)
    sam1N=np.array(np.random.choice(sam1, size=samSize, replace=True))
    sam2N=np.array(np.random.choice(sam2, size=samSize, replace=True))
    sam1Mean[i]=np.mean(sam1N)
    sam2Mean[i]=np.mean(sam2N)
  a=(1-CL)/NumTails
  z=stats.norm.ppf(1-a)

  mean1=np.round(np.mean(sam1Mean),2)
  sigma1=np.round(np.std(sam1Mean),2)
  lowLimit1=np.round(mean1-(z*sigma1),2)
  upLimit1=np.round(mean1+(z*sigma1),2)

  mean2=np.round(np.mean(sam2Mean),2)
  sigma2=np.round(np.std(sam2Mean),2)
  lowLimit2=np.round(mean2-(z*sigma2),2)
  upLimit2=np.round(mean2+(z*sigma2),2)

  f,a=plt.subplots(figsize=(20,5))
  san.histplot(data=sam1Mean, color="Red", kde=True)
  san.histplot(data=sam2Mean, color="Green", kde=True)
  label_mean1=("Mean(Male)", mean1)
  label_Llimit1=("Lower limit(Male)",lowLimit1)
  label_Ulimit1=("Upper limit(Male)", upLimit1 )
  print(f' The confidence interval range from {label_Llimit1} to {label_Ulimit1} and mean {label_mean1}')

  label_mean2=("Mean(Female)", mean2)
  label_Llimit2=("Lower limit(Female)",lowLimit2)
  label_Ulimit2=("Upper limit(Female)", upLimit2 )
  print(f' The confidence interval range from {label_Llimit2} to {label_Ulimit2} and mean {label_mean2}')
  plt.xlabel("Purchase")
  plt.title(f'Sample size: {samSize}, Male Avg: {mean1}, Male stdDev: {sigma1},Female Avg: {mean2}, Female stdDev: {sigma2}')
  plt.show()

  return sam1Mean, sam2Mean, np.round(lowLimit1,2), np.round(upLimit1,2),np.round(lowLimit2,2), np.round(upLimit2,2)
bootstrapMvsF(dfMale, dfFemale, samSize=100, itrSize=5000, CL=0.95, NumTails=2)
bootstrapMvsF(dfMale, dfFemale, samSize=1000, itrSize=5000, CL=0.95, NumTails=2)
bootstrapMvsF(dfMale, dfFemale, samSize=10000, itrSize=5000, CL=0.95, NumTails=2)
```

```python
def bootstrapMvsUM(sam1, sam2, samSize=500, itrSize=5000, CL=0.95, NumTails=2):
  sam1Mean=np.empty(itrSize)
  sam2Mean=np.empty(itrSize)
  for i in range(itrSize):
    sam1N=np.empty(samSize)
    sam2N=np.empty(samSize)
    sam1N=np.array(np.random.choice(sam1, size=samSize, replace=True))
    sam2N=np.array(np.random.choice(sam2, size=samSize, replace=True))
    sam1Mean[i]=np.mean(sam1N)
    sam2Mean[i]=np.mean(sam2N)
  a=(1-CL)/NumTails
  z=stats.norm.ppf(1-a)

  mean1=np.round(np.mean(sam1Mean))
  sigma1=np.round(np.std(sam1Mean))
  lowLimit1=np.round(mean1-(z*sigma1))
  upLimit1=np.round(mean1+(z*sigma1))
  mean2=np.round(np.mean(sam2Mean))
  sigma2=np.round(np.std(sam2Mean))
  lowLimit2=np.round(mean2-(z*sigma2))
  upLimit2=np.round(mean2+(z*sigma2))

  f,a=plt.subplots(figsize=(20,5))
  san.histplot(data=sam1Mean, color="Red")
  san.histplot(data=sam2Mean, color="Green")
  label_mean1=("Mean(Married)", mean1)
  label_Llimit1=("Lower limit(Married)",lowLimit1)
  label_Ulimit1=("Upper limit(Married)", upLimit1 )
  print(f' The confidence interval range from {label_Llimit1} to {label_Ulimit1} and mean {label_mean1}')
  label_mean2=("Mean(Unmarried)", mean2)
  label_Llimit2=("Lower limit(Unmarried)",lowLimit2)
  label_Ulimit2=("Upper limit(Unmarried)", upLimit2 )
  print(f' The confidence interval range from {label_Llimit1} to {label_Ulimit1} and mean {label_mean1}')
  plt.title(f'Sample size: {samSize}, Male Avg: {mean1}, Male stdDev: {sigma1},Female Avg: {mean2}, Female stdDev: {sigma2}')
  plt.xlabel("Purchase")
  plt.show()
  return sam1Mean, sam2Mean, np.round(lowLimit1,2), np.round(upLimit1,2),np.round(lowLimit2,2), np.round(upLimit2,2)
bootstrapMvsUM(dfMar, dfUmar, samSize=100, itrSize=5000, CL=0.85, NumTails=2)
bootstrapMvsUM(dfMar, dfUmar, samSize=1000, itrSize=5000, CL=0.85, NumTails=2)
bootstrapMvsUM(dfMar, dfUmar, samSize=10000, itrSize=5000, CL=0.85, NumTails=2)
```
